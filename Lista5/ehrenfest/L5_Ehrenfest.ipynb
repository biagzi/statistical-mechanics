```python
from matplotlib import pyplot as plt
import random as rand
```

```python
def urna_T(N,T):
    N1 = N
    N2 = N - N1
    n1 = [N1]
    n2 = [N2]
    for t in range(T):
        r = rand.uniform(0,1)
        if (r <= N1/N):
            N1 = N1 - 1
            N2 = N2 + 1
        else:
            N1 = N1 + 1
            N2 = N2 - 1
        
        n1.append(N1)
        n2.append(N2)
    
    return n1,n2
```

```python
#Parâmetros
N = 200 #200
T = 2000

n1_20,n2 = urna_T(20,T)
n1_200,n2 = urna_T(200,T)
n1_2000,n2 = urna_T(2000,T)

t = [i for i in range(T+1)]

plt.plot(t,n1_20,'-',label="N=20")
plt.plot(t,n1_200,'-',label="N=200")
#plt.plot(t,n1_2000,'-',label="N=2000")
plt.hlines(100,0,T,'orange')
plt.hlines(10,0,T,'blue')
#plt.hlines(1000,0,T,'green')
plt.xlabel('Tempo')
plt.ylabel('N1')
plt.ylim(0,N+1)
plt.xlim(0,T+1)
plt.legend()
plt.title('Número de bolas na caixa 1 com o tempo')
```
